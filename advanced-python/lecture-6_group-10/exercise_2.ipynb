```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Exercise 2 : Attributes of top songs on Spotify
```

```python
df = pd.read_csv('spotify.csv')
df.dtypes
```

```python
plt.scatter(x = df.energy, y=df.valence, c=df.liveness, cmap='viridis', 
            s=100, alpha=0.5)
plt.title('Attributes of top 50 songs on Spotify in 2019')
plt.xlabel('Energy')
plt.ylabel('Valence')
plt.colorbar(label='Liveness')
```
