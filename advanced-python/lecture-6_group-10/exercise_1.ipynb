```python
# Exercise 1 : Titanic
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style='white')
```

```python
# 1 (a)
df = pd.read_csv('titanic.csv')
```

```python
df_class = df[df.Pclass > 1]
df_class.Pclass = df_class.Pclass.astype('str')
```

```python
sns.set_color_codes()
hist_plot = sns.histplot(data=df_class, x='Fare', hue='Pclass',
                hue_order=["3", "2"], stat='probability', kde=True, palette=['red','green']
            )
hist_plot.set_title('Titanic fare price by class')
hist_plot.set_ylabel('Proportion of passengers')
hist_plot.legend(['Second class', 'Third class'])
```

```python
# 1 (b)
scatter_plot = sns.scatterplot(data=df, x='Age', y='Fare', 
                               hue='Survived',palette=['red','green'], alpha=0.5)
scatter_plot.set_title('Passenger outcome by fare-price and age')
scatter_plot.set_ylim([0,270])
```
