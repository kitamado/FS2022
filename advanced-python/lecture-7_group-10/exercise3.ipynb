### 3. Weather data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
df = pd.read_csv('uk_weather.csv')
```

#### (a) Datetime 

```python
df['year'] = df['year'].fillna(0)
df['month'] = df['month'].fillna(0)
df['datetime'] = df.apply(lambda row: datetime.strptime(f"{int(row.year)}-{int(row.month)}", '%Y-%m') if (row.year!=0 and row.month!=0) else np.nan, axis=1)
df
```

#### (b) Datatype

```python
df.dtypes
# the datetime dtype is datetime    datetime64[ns]
```

#### (c) User query and response part 1

```python
from dateutil import parser
```

```python
station = input('Station > ').lower()
date_begin = parser.parse(input('From date > '), default=datetime(1978, 1, 1, 0, 0)) 
date_end = parser.parse(input('To date > '), default=datetime(1978, 1, 31, 0, 0)) 
mask0 = df['station'] == station
mask1 = df['datetime'] >= date_begin
mask2 = df['datetime'] <= date_end
mean_rainfall = df[mask0 & mask1 & mask2]['rain'].mean()
print('M Rainfall:', mean_rainfall)
```

#### (d) User query and response part 2

```python
station = input('Station > ').lower()
date_begin = parser.parse(input('From date > '), default=datetime(1978, 1, 1, 0, 0))
date_end = parser.parse(input('To date > '), default=datetime(1978, 1, 31, 0, 0)) 
mask0 = df['station'] == station
mask1 = df['datetime'] >= date_begin
mask2 = df['datetime'] <= date_end
df_reqs = df[mask0 & mask1 & mask2]
print('M Rainfall:',df_reqs['rain'].max())
time = df_reqs["datetime"].iloc[-1] - df['datetime'].iloc[df_reqs['rain'].idxmax()].to_pydatetime()
print('Days since maximum rainfall in this period:',time.days + 1)
```
